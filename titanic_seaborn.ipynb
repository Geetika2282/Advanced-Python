```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
titanic.head(500)
# titanic.size
```

```python
# Create a count plot for survival
plt.figure(figsize=(8, 6))  # Set the figure size
sns.countplot(x='survived', data=titanic)
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers Who Survived vs. Did Not Survive')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='age', y='survived', data=titanic)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='fare', y='class', data=titanic)
plt.show()
```

```python
sns.histplot(data=titanic, x='age', bins=20)
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))

# Histogram
sns.histplot(data=titanic, x='age', bins=20, ax=axs[0, 0])
axs[0, 0].set_title('Histogram of Age')

# Boxplot
sns.boxplot(data=titanic, x='age', y='survived', ax=axs[0, 1])
axs[0, 1].set_title('Boxplot of Age by Survived')

# Violin plot
sns.violinplot(data=titanic, x='fare', y='class', ax=axs[1, 0])
axs[1, 0].set_title('Violin Plot Fare by Class')

# Countplot
sns.countplot(data=titanic, x='survived', ax=axs[1, 1])
axs[1, 1].set_title('Countplot of Survival')

plt.show();
```
