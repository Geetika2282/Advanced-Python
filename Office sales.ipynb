```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
plt.figure(figsize=(10, 4), dpi=200) # PI (dots per inch) controls the number of pixels per inch at which a figure is rend
sns.countplot(x='division', data=df)
plt.show()
```

```python
# breakdown with another category with 'hue'
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(x='level of education', data=df, hue='training level')
plt.show()
```

```python
# Using palette
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(x='level of education', data=df, hue='training level', palette='Set1')
plt.show()
```

```python
# Using palette
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(x='level of education', data=df, hue='training level', palette='Paired')
plt.show()
```
