```python
# Define the age bins (ranges) and the corresponding labels
bins = [0, 18, 59, 100]  # Age ranges: 0-12 (Child), 13-59 (Adult), 60+ (Senior)
labels = ['Child', 'Adult', 'Senior']

# Create a new column 'age_group' in the dataset
titanic['age_group'] = pd.cut(titanic['age'], bins=bins, labels=labels)

# Display the first few rows to check the new 'age_group' column
print(titanic[['age', 'age_group']].head(50))
```

# Survival analysis

```python
# What is the overall survival rate (percentage of passengers who survived)?
total_pass = len(titanic)
survived = titanic[titanic['alive']=='yes'].shape[0]
survival_rate = (survived/ total_pass) * 100

print(f"Survival rate: {survival_rate:.2f}")
```

```python
# Calculate the number of survivors by gender
survived_by_gender = titanic[titanic['alive'] == 'yes']['sex'].value_counts()
survived_by_gender
```

```python
# How does survival vary by passenger class? (Were first-class passengers more likely to survive?)
survived = titanic[titanic['alive']=='yes']['class'].value_counts()
survived
```

```python
# How does the age affect survival chances? (Did children have a higher survival rate than adults?)
titanic[titanic['alive']=='yes']['age_group'].max()
```

```python
# What is the survival rate for passengers who embarked from different ports (C, Q, S)
titanic[titanic['alive']=='yes']['embarked'].value_counts()
```

```python
# Did traveling with family (siblings/spouses, parents/children) impact survival chances?
print("Didn't travel alone: ")
titanic[(titanic['alone'] == False) & (titanic['alive'] == 'yes')].shape[0]
```

# Relationships Between Variables

```python
# Is there any correlation between age and fare paid? Do younger or older passengers tend to pay more?
childFare = titanic[titanic['age_group']=='Child']['fare']
adultFare = titanic[titanic['age_group']=='Adult']['fare']
seniorFare = titanic[titanic['age_group']=='Senior']['fare']
```

```python
maxFare = [childFare.max(), adultFare.max(), seniorFare.max()]
print("Maximum fares:", maxFare)
print("Overall max fare:", max(maxFare))
```

```python
age_fare_correlation = titanic[['age', 'fare']].corr().iloc[0, 1]
print(f"Correlation between age and fare: {age_fare_correlation:.2f}")
```

```python
# What is the relationship between class and fare? (Do higher-class passengers pay more on average?)
avg_fare_by_class = titanic.groupby('pclass')['fare'].mean()
avg_fare_by_class
```

```python
# How does family size (calculated from the sibsp and parch columns) affect survival?
titanic['family_size'] = titanic['sibsp'] + titanic['parch']
```
