```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
y = 2*x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)
axes.plot(x, y)
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
a
```

```python
b
```

```python
# create an empty canvas
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b)
plt.show()
```

```python
# ################## Sub plots ################
fig, axes = plt.subplots(nrows = 2,ncols=2,figsize=(12,8))
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)
fig.subplots_adjust(left=None,bottom=None,right=None,top=None,wspace=0.9,hspace=0.1)
```

```python
# generate x-axis values(1 to 9)
x = np.arange(1,10)
```

```python
# generate y-axis values(1 to 18)
y = np.arange(1,19)
```

```python
# select the first 9 elements from y to match x-axis length 
y = y[:9] # slicing to get elements from index 0(inclusive) to 9(exclusive)
```

```python
# create a line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot Title Here')
plt.scatter(x,y,s=50,c='red') # increase marker size to 50
plt.savefig('example.png')
```
