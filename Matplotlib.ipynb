```python
import pandas as pd
import numpy as np
# Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# matplotlib is a low level graph plotting library in python that serves as a visualization utility
import matplotlib
print(matplotlib.__version__)
```

```python
x = np.array([0, 6])
y = np.array([0, 250])
# The plot() function is used to draw points (markers) in a diagram.
# By default, the plot() function draws a line from point to point.
plt.plot(x, y)
plt.show()
```

```python
#  (number of points must be same)
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])
plt.plot(x, y)
plt.show()
```

```python
# to plot only the points
x = ([2, 8])
y = ([1, 4])
# mfc = marker face color; mec = marker edge color
plt.plot(x, y, marker='o', mec='blue', mfc = 'hotpink', ms = '12') # if only 'o' written without 'marker' will get only points
plt.show()
```

```python
'''The line style can be written in a shorter syntax:
    . linestyle can be written as ls.
    . dotted can be written as :.
    . dashed can be written as --. '''
ypoints = np.array([3, 8, 1, 10])
# x taken by default
plt.plot(ypoints, linestyle = 'dotted', color='hotpink', linewidth = '5.5')
plt.show()
```

```python
# multiple lines

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
# here x1, x2 are taken by default (0, 1, 2, 3)
plt.plot(y1)
plt.plot(y2)

# create labels
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# create title
plt.title("Sports Watch Data", loc='left')

# adding grid lines
plt.grid()

plt.show()
```

```python
# display multiple plots: takes 3 arguements
# 1st and 2nd arguement: rows and columns; 3rd is: the plot of current graph
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)
plt.title('Shop 1')

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 1, 2)
plt.plot(x,y)
plt.title('Shop 2')

plt.subplots_adjust(hspace=0.5)
plt.suptitle("MY SHOP")
plt.show()
```

## Scatter plot

```python
# Scatter plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
# one: 
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color='hotpink')

# two: 
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color='blue')

plt.show()
```

```python
# color map

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

# adjust transparency of dots with 'alpha'
plt.scatter(x, y, c=colors, cmap='viridis', alpha=0.5)

plt.colorbar()

plt.show()
```

## Bars

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.title('Toys')

plt.bar(x,y)
# plt.show()

x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.subplot(1, 2, 2)
plt.title('Fruits')

plt.bar(x, y)

plt.suptitle('Bar plots')
plt.subplots_adjust(wspace=0.5)
# plt.subplots_adjust()

plt.show()
```

```python
# Horizontal bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, color='hotpink', height=0.5)
plt.show()
```

```python
# Horizontal bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color='hotpink', width=0.6)
plt.show()
```

## Pie charts

```python
y = np.array([35, 25, 25, 5])
plt.pie(y)
plt.show()
```

```python
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
y = np.array([35, 25, 25, 5])
plt.pie(y, labels = mylabels)
plt.show()
```

```python
# explode
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
y = np.array([35, 25, 25, 5])
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode=myexplode)
plt.legend(title="My fruits")
plt.show()
```
