```python
# 5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.
x1 = [10, 20, 30]
y1 = [40, 10, 30]
plt.plot(x1, y1)

x2 = [10, 20, 30]
y2 = [20, 40, 10]
plt.plot(x2, y2)
plt.legend(['line 1', 'line 2'])
plt.show()
```

```python
# 6. plot 2 or more line with legends, different width, colors
x1 = [10, 20, 30]
y1 = [40, 10, 30]
plt.plot(x1, y1, color='skyblue', linewidth=2)

x2 = [10, 20, 30]
y2 = [20, 40, 10]
plt.plot(x2, y2, color='hotpink', linewidth=0.9)
plt.legend(['line 1', 'line 2'])
plt.show()
```

```python
# 7. plot 2 or more lines with different styles.
x1 = [10, 20, 30]
y1 = [40, 10, 30]
plt.plot(x1, y1, linestyle = 'dashed', linewidth='3')

x2 = [10, 20, 30]
y2 = [20, 40, 10]
plt.plot(x2, y2, linestyle ='dotted')
plt.legend(['line 1', 'line 2'])
plt.show()
```

```python
# 8. plot 2 or more lines and set line markers.
x = [1,4,5,6,7]
y = [2,6,3,6,3]

plt.plot(x, y,linestyle='dashdot',marker='o',markerfacecolor='blue',markersize=10,linewidth='3',color='red')

plt.show()
```

```python
# 9. to display current axis limits values and set new axis values
import matplotlib.pyplot as plt
X = range(1, 50)
Y = [value * 3 for value in X]
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line.')
# shows the current axis limits values
print(plt.axis()) 
# set new axes limits
# Limit of x axis 0 to 100
# Limit of y axis 0 to 200
plt.axis([0, 100, 0, 200]) 
# Display the figure.
plt.show()
```
