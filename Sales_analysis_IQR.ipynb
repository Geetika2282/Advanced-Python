```python
import numpy as np
import pandas as pd
```

```python
sales = pd.read_csv('sales_data.csv')
sales
```

```python
df = pd.DataFrame(sales)
df
```

```python
df['Total_Sales'] = df['Quantity'] * df['Price_per_Unit']
```

```python
df
```

```python
newSales = df.to_csv('newSales.csv', index=False)
```

```python
df
```

```python
sales_by_product = df.groupby('Product')['Total_Sales'].sum()
sales_by_product
```

```python
max_sales_product = sales_by_product.idxmax()
max_sales_value = sales_by_product.max()
print(f"Highest sales product: {max_sales_product} with sales of {max_sales_value}")
```

```python
best_sellingRow = df.loc[df['Quantity'].idxmax()]
best_selling = best_sellingRow['Product']
print(f"The best selling product is: {best_selling}")
```

```python
q1 = df['Price_per_Unit'].quantile(0.25)
q3 = df['Price_per_Unit'].quantile(0.75)
IQR = q3 - q1
IQR
```

```python
lowerFence = q1 - 1.5*(IQR)
higherFence = q3 + 1.5*(IQR)
```

```python
outliers = df[(df['Price_per_Unit'] < lowerFence) | (df['Price_per_Unit'] > higherFence)]
outliers
```

```python
df['rank_qty_asc'] = df['Quantity'].rank()
df
```

```python
df['rank_qty_desc'] = df['Quantity'].rank(ascending=False)
```

```python
df
```

```python
# Replace outliers with median
median = df['Price_per_Unit'].median()
df['Price_per_Unit'] = df['Price_per_Unit'].apply(lambda x: median if x < lowerFence or x > higherFence else x)
```
