```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import pandas as pd
import numpy as np
```

```python
mos = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
product = ['A', 'B', 'C', 'D', 'E']
a = np.random.randint(50, 100, size=(12, 5))
df = pd.DataFrame(a, columns=product, index=mos)
```

```python
df
```

```python
# figsize creates a new graph figure or window for plotting
# if not given considers default size

plt.figure(figsize=(12, 7))

# Plotting line graphs
plt.plot(mos, df['A'], label='Product A', color='blue', marker='o', linestyle='-')
plt.plot(mos, df['B'], label='Product B', color='red', marker='s', linestyle='-')
plt.plot(mos, df['C'], label='Product C', color='green', marker='o', linestyle='-')
plt.plot(mos, df['D'], label='Product D', color='pink', marker='s', linestyle='-')
plt.plot(mos, df['E'], label='Product E', color='orange', marker='o', linestyle='-')
# Adding labels and title
plt.title('Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
```
