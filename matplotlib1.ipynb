```python
from matplotlib import pyplot as plt
import numpy as np

# Simple plot example
plt.plot([1, 2, 3], [4, 5, 8])
plt.title('Practice')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
 
```

```python
plt.plot([1, 2, 3], [1, 4, 9], label="Squares")
plt.plot([1, 2, 3], [1, 8, 27], label="Cubes")
plt.title("Squares and Cubes")
plt.legend()
```

# Pie chart

```python
areas = ['A', 'B', 'C', 'D', 'E']
budget = [2.5, 4, 10, 5, 6]
plt.pie(budget, labels=areas, autopct="%0.1f%%")
plt.show()
```

```python
x = np.random.randint(1, 100, 10)
y = np.random.randint(50, 150, 10)
plt.scatter(x, y)
plt.xlabel('Age')
plt.ylabel('Salary')
```

```python
pdtID = ['A', 'B', 'C', 'D', 'E']
price = [23, 45, 78, 90, 44]
plt.pie(price, labels=pdtID, autopct="%0.2f%%")
plt.scatter(pdtID, price)
plt.show()
```

```python
plt.plot(pdtID, price, color='pink', marker='*', markersize=10, linestyle='--')
plt.xlabel('Product ID')
plt.ylabel('Price')
plt.title('Product Price Comparison')
```

```python
pdtID = ['A', 'B', 'C', 'D', 'E']
onlinePrice = [23, 45, 78, 90, 44]
offlinePrice = [74, 15, 84, 49, 24]
dupliPrice = [19, 34, 46, 50, 46]
plt.plot(pdtID, onlinePrice, color='red', label='online price')
plt.plot(pdtID, offlinePrice, color='green', label='offline price')
plt.plot(pdtID, dupliPrice, color='blue', label='duplicate price')
plt.legend()
plt.show()
```
