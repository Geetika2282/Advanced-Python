# Barplot

```python
plt.figure(figsize=(12, 8))
res = sns.barplot(x='Product', y='Quantity', data=data, palette='Set2')
plt.show()
```

```python
sns.countplot(x='Product', data=data)
plt.show()
```

```python
sns.countplot(x='Product', hue='Quantity', data=data, palette='Set1')
plt.show()
```

```python
dataNew = {
     'Quantity': data['Quantity'],
    'Price': data['Price_per_Unit']
}
dataN = pd.DataFrame(dataNew)
dataN
```

```python
sns.heatmap(dataN.corr(), cbar=True, linewidths=0.5)
plt.show()
```
