```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('airline_passenger_satisfaction.csv')
df.head()
```

```python
sat = df['On-board Service']
```

```python
mean_sat = sat.mean()
median_sat = sat.median()
skewness_sat = sat.skew()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(sat,kde=True,color='skyblue',bins=30)
plt.axvline(mean_sat,color='red',linestyle='--',label=f'Mean:{mean_sat:.2f}')
plt.axvline(median_sat,color='black',linestyle='--',label=f'Median:{median_sat:.2f}')
plt.title(f'Distribution of satisfaction of customers\n Skewness:{skewness_sat:.2f}')
plt.xlabel('on board servive')
plt.ylabel('frequency')
plt.legend()
plt.tight_layout()
plt.show()
```
