```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SP500 oil gold bitcoin.csv')
df.head()
```

```python
df_BITCOIN = df['BITCOIN']
df_Gold = df['Gold']
```

```python
# Calculate correlation
# Correlation measures both the strength and direction of the linear relationship between two variables
# Correlation values are standardized.
print("Correlation: ")
corr = df[['BITCOIN', 'Gold', 'S&P500']].corr()
corr
```

```python
# Calculate covariance
# Covariance indicates the direction of the linear relationship between variables.
print("Covariance: ")
cov = df[['BITCOIN', 'Gold', 'S&P500']].cov()
cov
```

```python
# Data plot
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
axes[0].scatter(df['BITCOIN'], df['Gold'], color='blue', alpha=0.5)
axes[0].set_title(f"Correlation: {corr.loc['BITCOIN', 'Gold']:.2f}")
axes[0].grid(True)
```
